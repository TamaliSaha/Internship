### Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.
This task will be done in following steps:
1. First get the webpage https://www.naukri.com/
2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the
location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.

```python
#First connect to the driver

driver= webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver_win32\chromedriver.exe")

#open the naukri page in automated chrome browser

driver.get("https://www.naukri.com/")
```

```python
# Enter required designation and location for Data Analyst

#Enter “Data Analyst” in “Skill, Designations, Companies” field

des= driver.find_element(By.XPATH, "/html/body/div[1]/div[6]/div/div/div[1]/div/div/div/input")
des.send_keys("Data Analyst")

# Enter “Bangalore” in “enter the location” field

loc= driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input")
loc.send_keys("Bangalore")
```

```python
#click the search button

search= driver.find_element(By.CLASS_NAME, "qsbSubmit")
search.click()
```

```python
# lets make empty list to store data with the given conditions

title= []
location=[]
company=[]
experience=[]
```

```python
# Scraping first 10 job title from the given page

title_tags= driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')
for i in title_tags[0:10]:
    x= i.text
    title.append(x) 

# Scraping first 10 job location from the given page

loc_tags= driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi location"]')
for i in loc_tags[0:10]:
    x= i.text
    location.append(x)
    
# Scraping first 10 company name the given page

com_tags= driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in com_tags[0:10]:
    x= i.text
    company.append(x)

# Scraping first 10 experience from the given page

exp_tags=  driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi experience"]')
for i in exp_tags[0:10]:
    x= i.text
    experience.append(x)
```

```python
#checking the length

print (len(title), len(location), len(company), len(experience))
```

```python
# Creating DataFrame to scrap first 10 data

data_list= list(zip(title, location, company,experience ))

data= pd.DataFrame(data_list, 
    columns =['Job_Title', 'Job_Location', 'Company_Name', 'Experience_Required'],
    index=pd.RangeIndex(start=1, stop=(len(title)+1), name='Sr. No.'))

print ("First 10 Data Analyst jobs data from Naukri is:\n")
data
```

```python
#click on search 

search= driver.find_element(By.XPATH, "/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button")
search.click()
```

```python
#click the first results from all iphone 11 data

click_first = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[2]/div/div/div/a/div[2]')
click_first.click()
```

```python
# as the result open up in new tab of the browser we need to go that tab

driver.switch_to.window(driver.window_handles[1])
```

```python
#clicking on the all reviews button

all_rev = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[6]/div/a/div/span')
all_rev.click()
```

```python
#Scrap the required data Rating, Review summary, Full review

ratings=[]
review_summary=[]
full_review=[]
```

```python
# Scraping first 100 Ratings the given page

start = 0 #starting page
end = 10 # page number to end scaping


for page in range(start,end):
    
    #scrape the ratings
    
    rating_tags = driver.find_elements(By.XPATH,'//div[@class="_3LWZlK _1BLPMq"]')
    for i in rating_tags:
        x= i.text
        ratings.append(x)
        
    #getting the nextpage button & clicking next 
    
    next_button = driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    next_button.click()
    time.sleep(5)
```

```python
len(ratings)
```

```python
# Scraping first 100 Review summary the given page

start = 0 #starting page
end = 10 # page number to end scaping


for page in range(start,end):
    
    # scrape the review summaries
    
    review_sum_tags = driver.find_elements(By.XPATH,'//p[@class="_2-N8zT"]')
    for i in review_sum_tags:
        x= i.text
        review_summary.append(x)    
    
    #getting the nextpage button & clicking next 
    
    next_button = driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    next_button.click()
    time.sleep(5)
```

```python
len(review_summary)
```

```python
# Scraping first 100 full review the given page

start = 0 #starting page
end = 10 # page number to end scaping


for page in range(start,end):
    
    # scrape the full reviews
    
    full_review_tags = driver.find_elements(By.XPATH,'//div[@class="t-ZTKy"]')
    for i in full_review_tags:
        x= i.text
        full_review.append(x.replace("\n"," "))
        
    #getting the nextpage button & clicking next 
    
    next_button = driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    next_button.click()
    time.sleep(5)
```

```python
len(full_review)
```

```python
# Creating DataFrame to scrap first 10 data

data_list= list(zip(ratings, review_summary, full_review))

data= pd.DataFrame(data_list,
    columns =['Ratings', 'Review Summary', 'Full Summary'],
    index=pd.RangeIndex(start=1, stop=101, name='Sr. No.'))

data = data.style.set_properties(**{'text-align': 'right'})

#Print the required data

print ("First 100 reviews data from flipkart.com for iphone11 phone:\n")
data
```

### Q10: Write a python program to scrape the salary data for Data Scientist designation. You have to scrape Company name, Number of salaries, Average salary, Minsalary, Max Salary. 

The above task will be, done as shown in the below steps:
1. First get the webpage https://www.ambitionbox.com/
2. Click on the salaries option as shown in the image.
3. After reaching to the following webpage, In place of “Search Job Profile” enters “Data Scientist” and then click on “Data Scientist”.
4. Scrape the data for the first 10 companies. Scrape the company name, total salary record, average salary, minimum salary, maximum salary, experience required.
5. Store the data in a dataframe.


```python
#Lets first connect to the driver

driver= webdriver.Chrome(r"chromedriver.exe")

#Go to ambitionbox webpage by url:  https://www.ambitionbox.com/

driver.get("https://www.ambitionbox.com/")
```

```python
# click to close popup window

popup = driver.find_element(By.XPATH,'/html/body/div/div/div/div[4]/div/div/button')

popup.click()
```

```python
# Click on the salaries option

sal_click = driver.find_element(By.XPATH,'/html/body/div/div/div/div[1]/header/nav/ul/li[3]')

sal_click.click()
```

```python
# click the browse salaries button 

browse_sal = driver.find_element(By.PARTIAL_LINK_TEXT, 'Salaries')

browse_sal.click()
```

```python
# Enters “Data Scientist” in place of “Search Job Profile” 

data_sc= driver.find_element(By.XPATH,'/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/input')

data_sc.send_keys('Data Scientist')
```

```python
# click the on “Data Scientist”

data_sc_click = driver.find_element(By.XPATH,'/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/div/div/div[1]/div/div/p')

data_sc_click.click()
```

```python
#Scrap the company name, total salary record, average salary, minimum salary, maximum salary, experience required

company=[]
total_salary=[]
avg_salary=[]
min_salary=[]
max_salary=[]
exp=[]
```

```python
# Scrape the data for the first 10 companies

#to get company name

company_tags = driver.find_elements(By.XPATH,'//a[@data-v-4c07f399]')
for i in company_tags:
    x= i.text
    company.append(x.replace('\nData Scientist Salary'," "))

#to get company name

tol_sal_tags = driver.find_elements(By.XPATH,'//span[@class="datapoints"]')
for i in tol_sal_tags:
    x= i.text[10:12]
    total_salary.append(x)
    
#to get avg salary

avg_sal = driver.find_elements(By.XPATH,'//p[@class="averageCtc"]')
for i in avg_sal:
    x= i.text
    avg_salary.append(x)

#to get minimum salary

min_sal_tags = driver.find_elements(By.XPATH,'//div[@class="salary-values"]')
for i in min_sal_tags:
    x= i.text
    min_salary.append(x.split('\n')[0])

#to get maximum salary

max_sal_tags = driver.find_elements(By.XPATH,'//div[@class="salary-values"]')
for i in max_sal_tags:
    x= i.text
    max_salary.append(x.split('\n')[1].strip())
    
#to get experience

exp_tags = driver.find_elements(By.XPATH,'//div[@class="sbold-list-header"]')
for i in exp_tags:
    x= i.text
    exp.append(x.split('e')[0])
```

```python
print(len(company), len(total_salary), len(avg_salary), len(min_salary), len(max_salary), len(exp))
```

```python
# Creating DataFrame to store the required field of first 10 data

data_list = list(zip(company,total_salary,avg_salary, min_salary,max_salary,exp))

data= pd.DataFrame(data_list, 
    columns =['Company Name', 'Total Salary','Average Salary','Minimum Salary','Maximum Salary', 'Experixence'],
    index=pd.RangeIndex(start=1, stop=11, name='Sr. No.'))

#print the outcomes

print ("First 10 Data Scientist salary data for from Ambitionbox.com is:\n\n")
data
```
